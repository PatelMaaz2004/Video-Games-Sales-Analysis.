### Handling Missing Values

```python
games.isnull().sum()
# To know the null values from the dataset.
```

#### Year_of_Release

```python
games['Year_of_Release'].fillna(games['Year_of_Release'].median(), inplace=True)
# we will fill with medianto preserve the timeline analysis.
```

```python
games['Year_of_Release'].isnull().sum()
# Now there are no missing values in this column.
```

```python
games['Publisher'].fillna('Unknown', inplace=True)
```

```python
games['Publisher'].isnull().sum()
```

#### Now the rest of the columns like:
1. Critic_Score
2. Critic_Count
3. User_Score
4. User_Count
5. Developer
6. Rating
#### They have a lot of missing values and also they are not valuable for us to do EDA.

```python
games.drop(columns=['Critic_Score','Critic_Count','User_Score','User_Count','Developer','Rating'],inplace=True)
```

```python
games.isnull().sum()
# we also have 2 more columns with 2 missing values in them. We will handle them as well.We will eliminate them.
```

```python
missing_rows = games[games['Name'].isnull() | games['Genre'].isnull()]
missing_rows
# to identify the rows with missing values.
```

```python
games.drop(index=missing_rows.index, inplace=True)
```

```python
games.isnull().sum()
```

#### Now we have no missing values in the dataset.

## Data Visualization

### 1. Most Common Games Types

```python
plt.figure(figsize=(10,5))
sns.countplot(data=games, y='Genre', order=games['Genre'].value_counts().index)
plt.title('Number of Games by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

### Action and Sports captures the most of the market of games as compared to rest of the games.

### 2. Number of games added per platform

```python
plt.figure(figsize=(12,5))
sns.countplot(data=games, x='Platform', order=games['Platform'].value_counts().index)
plt.title('Number of Games per Platform')
plt.xticks(rotation=45)
plt.show()
```

### PS2 and DS have the highest number of game releases, indicating longer platform and strong developer support.

### 3. Global Video Game Sales Trend Over Time

```python
yearly_sales = games.groupby('Year_of_Release')['Global_Sales'].sum()

plt.figure(figsize=(12,5))
yearly_sales.plot()
plt.title('Global Sales Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Global Sales (Millions)')
plt.show()
```

### Global video game sales show a steady rise from the early years, peaking around the late 2000s and then back to low in the year 2014 and then after.â€

### 4. Regional Sales Comparison

```python
region_sales = games[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()

plt.figure(figsize=(7,5))
region_sales.plot(kind='pie', autopct='%.2f%%')
plt.title('Regional Contribution to Global Sales')
plt.ylabel('')
plt.show()
```

### North America contributes the largest share of global sales.

### 

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=games['Global_Sales'])
plt.title('Outliers in Global Sales')
plt.show()
```

### 5. Platform VS Global Sales

```python
platform_sales = games.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))
platform_sales.head(10).plot(kind='bar')
plt.title('Top Platforms by Global Sales')
plt.ylabel('Global Sales (Millions)')
plt.show()
```

### PS2 contrinutes the highest global sales among all platforms, making it top revenue generating platform compared to other games.
